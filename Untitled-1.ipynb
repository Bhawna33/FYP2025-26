```python
def get_label(filename):
    f = filename.upper()

    # Left-hand motor imagery tasks
    if any(x in f for x in ["R03", "R07", "R11"]):
        return 0

    # Right-hand motor imagery tasks
    if any(x in f for x in ["R04", "R08", "R12"]):
        return 1

    return None   # ignore other runs
```

```python
def get_label(file_path):
    fname = file_path.split("/")[-1]

    # Label mapping
    label_map = {
        "R03": 0, "R07": 0, "R11": 0,   # Rest
        "R04": 1, "R08": 1, "R12": 1,   # Workload
    }

    # Extract Rxx
    for key in label_map.keys():
        if key in fname:
            return label_map[key]

    return None  # Not a workload/rest task file
```

```python
import mne
import numpy as np

fixed_len = 480   # 3 seconds at 160 Hz
X_all = []
y_all = []

for f in edf_files:
    label = get_label(f)
    if label is None:
        continue

    raw = mne.io.read_raw_edf(f, preload=True, verbose=False)
    data = raw.get_data()[:64, :]   # first 64 channels

    # If too long ‚Üí crop
    if data.shape[1] > fixed_len:
        data = data[:, :fixed_len]

    # If too short ‚Üí pad
    elif data.shape[1] < fixed_len:
        pad_width = fixed_len - data.shape[1]
        data = np.pad(data, ((0,0), (0, pad_width)), mode='constant')

    # Now shape is exactly (64, fixed_len)
    X_all.append(data)
    y_all.append(label)

X_all = np.array(X_all)
y_all = np.array(y_all)

print("Final X shape:", X_all.shape)
print("Final y shape:", y_all.shape)
print("Unique labels:", np.unique(y_all))
```

```python
# import numpy as np
# from scipy.signal import welch
# from numpy import trapz
# from scipy.stats import kurtosis, skew
# from sklearn.preprocessing import StandardScaler

# sfreq = 160  # sampling frequency from EEGMMIDB

# bands = {
#     "delta": (0.5, 4),
#     "theta": (4, 8),
#     "alpha": (8, 13),
#     "beta": (13, 30),
#     "gamma": (30, 45)
# }

# def spectral_entropy(psd, freqs):
#     """Normalized spectral entropy (base 2). psd and freqs are 1D arrays."""
#     psd = psd.copy()
#     psd = psd / (psd.sum() + 1e-12)
#     psd = psd + 1e-12  # avoid log(0)
#     se = -np.sum(psd * np.log2(psd))
#     # normalize by log2(number of bins)
#     return se / np.log2(len(psd))

# def hjorth_parameters(x):
#     """Return (activity, mobility, complexity) for 1D signal x."""
#     x = np.asarray(x)
#     dx = np.diff(x)
#     ddx = np.diff(dx)
#     var_x = x.var()
#     var_dx = dx.var() if dx.size else 0
#     var_ddx = ddx.var() if ddx.size else 0
#     activity = var_x
#     mobility = np.sqrt(var_dx / (var_x + 1e-12))
#     complexity = np.sqrt((var_ddx / (var_dx + 1e-12)) / (mobility**2 + 1e-12))
#     return activity, mobility, complexity

# def extract_bandpower_enhanced(eeg_sample, nperseg=256):
#     """
#     eeg_sample: (n_channels, n_times)
#     returns: 1D feature vector
#     """
#     n_ch = eeg_sample.shape[0]
#     band_names = list(bands.keys())

#     features = []

#     # precompute PSD for each channel
#     for ch in range(n_ch):
#         signal = eeg_sample[ch]
#         # Welch PSD
#         f, Pxx = welch(signal, fs=sfreq, nperseg=nperseg)

#         # total power in 0.5-45 Hz for normalization
#         total_mask = (f >= 0.5) & (f <= 45)
#         total_power = trapz(Pxx[total_mask], f[total_mask]) + 1e-12

#         # per-channel time-domain Hjorth
#         a, m, c = hjorth_parameters(signal)
#         features.extend([a, m, c])

#         # other time-domain stats (optionally)
#         features.append(np.mean(signal))
#         features.append(np.std(signal))
#         features.append(skew(signal))
#         features.append(kurtosis(signal))

#         # spectral descriptors
#         features.append(spectral_entropy(Pxx[total_mask], f[total_mask]))
#         # spectral centroid
#         centroid = np.sum(f[total_mask] * Pxx[total_mask]) / (np.sum(Pxx[total_mask]) + 1e-12)
#         features.append(centroid)

#         # per-band metrics
#         band_powers = []
#         for bname in band_names:
#             low, high = bands[bname]
#             mask = (f >= low) & (f <= high)
#             band_power = trapz(Pxx[mask], f[mask]) + 1e-12
#             band_powers.append(band_power)

#             # raw band power, relative band power, log-bandpower
#             features.append(band_power)
#             features.append(band_power / total_power)
#             features.append(np.log(band_power))

#             # peak freq within band
#             if mask.any():
#                 peak_idx = np.argmax(Pxx[mask])
#                 peak_freq = f[mask][peak_idx]
#             else:
#                 peak_freq = 0.0
#             features.append(peak_freq)

#         # band ratios (common discriminative ratios)
#         # e.g., alpha/beta, theta/alpha
#         # ensure no division by zero
#         alpha = band_powers[band_names.index("alpha")]
#         beta = band_powers[band_names.index("beta")]
#         theta = band_powers[band_names.index("theta")]
#         delta = band_powers[band_names.index("delta")]
#         gamma = band_powers[band_names.index("gamma")]
#         features.append(alpha / (beta + 1e-12))
#         features.append(theta / (alpha + 1e-12))
#         features.append((alpha + beta) / (delta + 1e-12))

#     # features is a list; return as numpy array
#     return np.array(features, dtype=np.float32)


# # Example usage to get X_features from X_all
# # X_all shape assumed (n_samples, n_channels, n_times)
# def batch_extract(X_all, nperseg=256, scaler=None):
#     feat_list = []
#     for x in X_all:
#         feat_list.append(extract_bandpower_enhanced(x, nperseg=nperseg))
#     Xf = np.vstack(feat_list)  # (n_samples, n_features)

#     # optional scaling
#     if scaler is not None:
#         Xf = scaler.transform(Xf)
#     return Xf



# import numpy as np
# from scipy.signal import welch
# from numpy import trapezoid  # correct function

# # Define EEG bands
# bands = {
#     "delta": (0.5, 4),
#     "theta": (4, 8),
#     "alpha": (8, 13),
#     "beta": (13, 30),
#     "gamma": (30, 45)
# }

# sfreq = 160   # EEG sampling frequency (eegmmidb)

# def extract_bandpower_features(eeg_sample):
#     """
#     eeg_sample shape ‚Üí (64, 480)
#     Converts EEG ‚Üí feature vector of size 64*5 bands = 320 features
#     """
#     features = []

#     for ch in range(eeg_sample.shape[0]):
#         f, Pxx = welch(eeg_sample[ch], fs=sfreq, nperseg=256)
#         for (low, high) in bands.values():
#             mask = (f >= low) & (f <= high)
#             band_power = trapezoid(Pxx[mask], f[mask])  # use trapezoid here
#             features.append(band_power)

#     return np.array(features)

# # Extract features for all samples
# X_features = np.array([extract_bandpower_features(x) for x in X_all])

# print("Original X:", X_all.shape)
# print("Feature X:", X_features.shape)
# print("Labels:", y_all.shape)

import numpy as np
from scipy.signal import welch
from numpy import trapz
from scipy.stats import kurtosis, skew

sfreq = 160  # PhysioNet EEG sampling frequency

bands = {
    "delta": (0.5, 4),
    "theta": (4, 8),
    "alpha": (8, 13),
    "beta": (13, 30),
    "gamma": (30, 45)
}

# -------------------------------
#    Helper Functions
# -------------------------------

def spectral_entropy(psd):
    psd = psd + 1e-12
    psd_norm = psd / np.sum(psd)
    return -np.sum(psd_norm * np.log2(psd_norm))

def hjorth_params(x):
    dx = np.diff(x)
    ddx = np.diff(dx)
    var0 = np.var(x)
    var1 = np.var(dx)
    var2 = np.var(ddx)

    activity = var0
    mobility = np.sqrt(var1 / (var0 + 1e-12))
    complexity = np.sqrt((var2 / (var1 + 1e-12)) / (mobility + 1e-12))
    return activity, mobility, complexity

def zero_crossings(x):
    return ((x[:-1] * x[1:]) < 0).sum()

def waveform_length(x):
    return np.sum(np.abs(np.diff(x)))

# -------------------------------
#    Main Extractor
# -------------------------------

def extract_features_custom(eeg_sample):
    """
    eeg_sample shape: (64, 480)
    returns ~2880 features
    """

    features = []

    for ch in range(eeg_sample.shape[0]):
        x = eeg_sample[ch]

        # PSD
        f, Pxx = welch(x, fs=sfreq, nperseg=256)
        total_power = trapz(Pxx, f) + 1e-12

        # Time-domain features
        activity, mobility, complexity = hjorth_params(x)
        features += [
            np.mean(x),
            np.std(x),
            np.median(x),
            skew(x),
            kurtosis(x),
            np.max(x) - np.min(x),   # peak-to-peak
            np.sqrt(np.mean(x**2)),  # RMS
            zero_crossings(x),
            waveform_length(x),
            activity,
            mobility,
            complexity
        ]

        # Frequency-domain features
        for bname, (low, high) in bands.items():
            mask = (f >= low) & (f <= high)
            band_power = trapz(Pxx[mask], f[mask]) + 1e-12

            features.append(band_power)                  # raw
            features.append(band_power / total_power)    # relative
            features.append(np.log(band_power))          # log-power

            # peak frequency in band
            if np.any(mask):
                peak_freq = f[mask][np.argmax(Pxx[mask])]
            else:
                peak_freq = 0.0
            features.append(peak_freq)

        # spectral descriptors
        features.append(spectral_entropy(Pxx))
        spectral_centroid = np.sum(f * Pxx) / (np.sum(Pxx) + 1e-12)
        features.append(spectral_centroid)

    return np.array(features, dtype=np.float32)

# Batch processing
def batch_extract_custom(X_all):
    return np.vstack([extract_features_custom(x) for x in X_all])

```

```python
import pyautogui
import time

def react_to_state(state):
    if state == 0:
        print("Relaxed ‚Üí Moving mouse slightly left")
        pyautogui.moveRel(-50, 0, duration=0.5)

    elif state == 1:
        print("Focused ‚Üí Moving mouse slightly right")
        pyautogui.moveRel(50, 0, duration=0.5)

    elif state == 2:
        print("Stress ‚Üí Moving mouse up")
        pyautogui.moveRel(0, -50, duration=0.5)

    elif state == 3:
        print("Drowsy ‚Üí Moving mouse down")
        pyautogui.moveRel(0, 50, duration=0.5)

    else:
        print("Unknown state, no action")
```
